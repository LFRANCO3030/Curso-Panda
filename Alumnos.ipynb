```python
datos = pd.read_csv(url)
datos['Puntos_extra'] = datos['Nota']*0.40
datos.head()
```

```python
datos['Notas_finales'] = datos['Puntos_extra'] + datos['Nota']
datos.head()
```

### Aprobado Final : True: si el estudiante est aprobado (la nota final debe ser mayor o igual a 7.0). False: si el estudiante est reprobado (la nota final debe ser menor que 7.0).

```python
datos['Aprobado Final'] = datos['Notas_finales'].apply(lambda x: 'True' if x > 7 else 'False')
datos.head()
```

```python
seleccion = (datos['Aprobado'] == False) & (datos['Aprobado Final'] == True)
datos[seleccion]
```
